```python
import pandas as pd
import numpy as np
import requests
import zipfile
import os
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns

class GravityModelEngine:
    def __init__(self, zip_url, fta_partners, start_year, end_year):
        self.zip_url = zip_url
        self.fta_partners = fta_partners
        self.start_year = start_year
        self.end_year = end_year
        self.data = None
        self.cleaned_data = None

    def download_and_extract_data(self, extract_to='./gravity_data'):
        print("Downloading and extracting data...")
        local_filename = self.zip_url.split('/')[-1]
        with requests.get(self.zip_url, stream=True) as r:
            with open(local_filename, 'wb') as f:
                f.write(r.content)
        with zipfile.ZipFile(local_filename, 'r') as zip_ref:
            zip_ref.extractall(extract_to)
        os.remove(local_filename)
        self.data_path = os.path.join(extract_to, 'Gravity_V202211.csv')

    def load_data(self):
        print("Loading data...")
        try:
            self.data = pd.read_csv(self.data_path)
            print("Data Headers:", self.data.columns.tolist())
            #print("First few rows:")
            #print(self.data.head())
        except Exception as e:
            print(f"Error loading data: {e}")

    def filter_and_clean_data(self):
        print("Filtering and cleaning data...")
        try:
            # Step 1: Filter for Azerbaijan and FTA partners
            self.cleaned_data = self.data[
                ((self.data['iso3_o'] == 'AZE') & (self.data['iso3_d'].isin(self.fta_partners))) |
                ((self.data['iso3_d'] == 'AZE') & (self.data['iso3_o'].isin(self.fta_partners)))
            ]
            print(f"After filtering for Azerbaijan and FTA partners: {self.cleaned_data.shape}")

            # Step 2: Filter for the year range
            self.cleaned_data = self.cleaned_data[
                (self.cleaned_data['year'] >= self.start_year) & (self.cleaned_data['year'] <= self.end_year)
            ]
            print(f"After filtering for years {self.start_year}-{self.end_year}: {self.cleaned_data.shape}")

            # Step 3: Drop rows with missing values in key columns
            required_columns = ['gdp_o', 'gdp_d', 'tradeflow_baci', 'dist', 'comlang_off', 'contig', 'fta_wto']
            print("Checking for missing values in required columns...")
            print(self.cleaned_data[required_columns].isnull().sum())
            self.cleaned_data = self.cleaned_data.dropna(subset=required_columns)
            print(f"After dropping rows with missing values in {required_columns}: {self.cleaned_data.shape}")

            # Step 4: Create log-transformed variables
            if not self.cleaned_data.empty:
                # Handle zero and missing values before log transformation
                self.cleaned_data['log_gdp_o'] = np.log(self.cleaned_data['gdp_o'].replace(0, np.nan))
                self.cleaned_data['log_gdp_d'] = np.log(self.cleaned_data['gdp_d'].replace(0, np.nan))
                self.cleaned_data['log_dist'] = np.log(self.cleaned_data['dist'].replace(0, np.nan))
                self.cleaned_data['log_tradeflow'] = np.log(self.cleaned_data['tradeflow_baci'].replace(0, np.nan) + 1)
                
                # Drop rows where log transformation introduced NaN
                self.cleaned_data.dropna(subset=['log_gdp_o', 'log_gdp_d', 'log_dist', 'log_tradeflow'], inplace=True)
                print(f"Data after log transformation: {self.cleaned_data.shape}")
            else:
                raise ValueError("Cleaned dataset is empty after filtering and cleaning.")

            print("Cleaned Data Summary:")
            print(self.cleaned_data.describe())
        except Exception as e:
            print(f"Error during filtering and cleaning: {e}")

    def visualize_trade_flows(self):
        if self.cleaned_data is None or self.cleaned_data.empty:
            print("No data available for visualization.")
            return

        print("Visualizing trade flows...")
        try:
            plt.figure(figsize=(10, 6))
            sns.lineplot(data=self.cleaned_data, x='year', y='tradeflow_baci', hue='iso3_d', marker="o")
            plt.title("Trade Flows Between Azerbaijan and FTA Partners Over Time", fontsize=14)
            plt.xlabel("Year", fontsize=12)
            plt.ylabel("Trade Flow (BACI)", fontsize=12)
            plt.legend(title="Partner Country", bbox_to_anchor=(1.05, 1), loc='upper left')
            plt.tight_layout()
            plt.grid(True)
            plt.show()
        except Exception as e:
            print(f"Error during visualization: {e}")

    def run_gravity_model(self):
        if self.cleaned_data is None or self.cleaned_data.empty:
            print("No data available for running the gravity model.")
            return None

        print("Running Gravity Model...")
        try:
            gravity_model = smf.ols(
                formula="log_tradeflow ~ log_gdp_o + log_gdp_d + log_dist + contig + fta_wto + comlang_off + col_dep + gdpcap_d",
                data=self.cleaned_data
            ).fit()
            print("Gravity Model Summary:")
            print(gravity_model.summary())
            return gravity_model
        except Exception as e:
            print(f"Error during gravity model estimation: {e}")
            return None

    def run_ppml_model(self):
        if self.cleaned_data is None or self.cleaned_data.empty:
            print("No data available for running the PPML model.")
            return None

        print("Running PPML Model...")
        try:
            ppml_model = smf.glm(
                formula="tradeflow_baci ~ log_gdp_o + log_gdp_d + log_dist + contig + fta_wto + comlang_off + col_dep + gdpcap_d",
                data=self.cleaned_data,
                family=sm.families.Poisson()
            ).fit()
            print("PPML Model Summary:")
            print(ppml_model.summary())
            return ppml_model
        except Exception as e:
            print(f"Error during PPML model estimation: {e}")
            return None

    def save_results(self, model, filename):
        if model is None:
            print("No model results to save.")
            return

        print(f"Saving results to {filename}...")
        try:
            results = pd.DataFrame(model.summary().tables[1].data)
            results.to_csv(filename, index=False)
        except Exception as e:
            print(f"Error saving results: {e}")

# Usage
if __name__ == "__main__":
    # Define parameters
    zip_url = "https://www.cepii.fr/DATA_DOWNLOAD/gravity/data/Gravity_csv_V202211.zip"
    fta_partners = ["RUS", "MDA", "UKR", "GEO", "TKM", "UZB", "KAZ", "KGZ", "BLR", "TJK"]
    start_year = 2000
    end_year = 2020

    # Initialize the analysis class
    engine = GravityModelEngine(zip_url, fta_partners, start_year, end_year)

    # Perform analysis
    engine.download_and_extract_data()
    engine.load_data()
    engine.filter_and_clean_data()
    engine.visualize_trade_flows()

    # Run gravity model
    gravity_model = engine.run_gravity_model()

    # Run PPML model
    ppml_model = engine.run_ppml_model()

    # Save results
    engine.save_results(gravity_model, "gravity_model_results.csv")
    engine.save_results(ppml_model, "ppml_model_results.csv")
```

```python
import pandas as pd
import numpy as np
import gme as gme
import gegravity as ge

# Load the gravity data
gravity_data_location = "/Users/toghruln/Downloads/azerbaijan_trade_data_cleaned.csv"
grav_data = pd.read_csv(gravity_data_location)
print(grav_data.columns)  # Check the column names in the dataset

# Ensure necessary columns are included in baseline_data
required_columns = ["year", "gdp_o", "gdp_d", "iso3_o", "iso3_d", "tradeflow_baci", 
                    "pop_o", "pop_d", "dist", "contig", "comlang_off", "fta_wto"]
if not all(col in grav_data.columns for col in required_columns):
    raise ValueError(f"The dataset is missing one or more required columns: {required_columns}")

# Fill missing values in critical columns
grav_data[required_columns] = grav_data[required_columns].fillna(0)

# Check for missing bilateral pairs and fill
country_set = set(grav_data["iso3_o"]).union(set(grav_data["iso3_d"]))
missing_pairs = [
    (exp, imp) for exp in country_set for imp in country_set
    if not ((grav_data["iso3_o"] == exp) & (grav_data["iso3_d"] == imp)).any()
]
for exp, imp in missing_pairs:
    grav_data = pd.concat([grav_data, pd.DataFrame([{
        "year": 2020,  # Default year
        "gdp_o": 0,
        "gdp_d": 0,
        "iso3_o": exp,
        "iso3_d": imp,
        "tradeflow_baci": 0,
        "pop_o": 0,
        "pop_d": 0,
        "dist": np.nan,
        "contig": 0,
        "comlang_off": 0,
        "fta_wto": 0
    }])])

# Fill missing cost variables with defaults
grav_data["dist"].fillna(grav_data["dist"].mean(), inplace=True)
grav_data["contig"].fillna(0, inplace=True)

# Create baseline_data with necessary columns
baseline_data = grav_data[required_columns].copy()

# Prepare the estimation data
gme_data = gme.EstimationData(
    grav_data,
    exp_var_name="iso3_o",
    imp_var_name="iso3_d",
    year_var_name="year",
    trade_var_name="tradeflow_baci"
)

# Define and estimate the model
gme_model = gme.EstimationModel(
    gme_data,
    lhs_var="tradeflow_baci",
    rhs_var=["contig", "dist", "comlang_off", "fta_wto"],
    fixed_effects=["iso3_o", "iso3_d"]
)
gme_model.estimate()
print(gme_model.results_dict["all"].summary())

# Build a general equilibrium model with corrected baseline type
baseline = ge.BaselineData(
    baseline_data,
    exp_var_name="iso3_o",
    imp_var_name="iso3_d",
    output_var_name="gdp_d",
    year_var_name="year",
    trade_var_name="tradeflow_baci"
)

# Extract cost coefficients in the format required by OneSectorGE
cost_coeff_values = gme_model.results_dict["all"]

# Initialize the GE model
ge_model = ge.OneSectorGE(
    baseline,
    year="2020",
    reference_importer="AZE",  # Reference set to Azerbaijan
    expend_var_name="gdp_o",
    output_var_name="gdp_d",
    sigma=5,
    cost_variables=["dist", "contig"],
    cost_coeff_values=cost_coeff_values
)

# Build the baseline for the GE model
ge_model.build_baseline(omr_rescale=10)
print(ge_model.baseline_mr)

# Update the FTA dummy variable for specific country pairs
exp_data = ge_model.baseline_data.copy()
fta_partners = ["RUS", "MDA", "UKR", "GEO", "TKM", "UZB", "KAZ", "KGZ", "BLR", "TJK"]
for country in fta_partners:
    exp_data.loc[(exp_data["iso3_d"] == "AZE") & (exp_data["iso3_o"] == country), "fta_wto"] = 1
    exp_data.loc[(exp_data["iso3_d"] == country) & (exp_data["iso3_o"] == "AZE"), "fta_wto"] = 1

# Supply the updated data for the experiment
ge_model.define_experiment(exp_data)

# Simulate the general equilibrium model
ge_model.simulate()

# Extract results
country_results = ge_model.country_results
bilateral_results = ge_model.bilateral_trade_results

# Export results
ge_model.export_results(directory="/Users/toghruln/Downloads", name="Azerbaijan_Results")
```
